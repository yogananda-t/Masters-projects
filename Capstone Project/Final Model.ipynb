### Handling the Outliers

```python
# Loop through numeric columns in df2 to detect outliers using the IQR method.
for column in df2.select_dtypes(include=['float64', 'int64']).columns:
     # Calculate the first quartile (Q1) and third quartile (Q3)
    Q1 = df2[column].quantile(0.25)
    Q3 = df2[column].quantile(0.75)
     # Compute Interquartile Range (IQR) and define bounds for outliers.
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Check for outliers
    outliers_condition = (df2[column] < lower_bound) | (df2[column] > upper_bound)
    if any(outliers_condition):
        print(f"Outliers detected in column: {column}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd



def melt_dataframe(df, columns):
    # Ensures that the 'columns' parameter is a list, for consistency
    if isinstance(columns, str):
        columns = [columns]
    # Melt the DataFrame based on the specified columns
    return df.melt(value_vars=columns, var_name='Columns', value_name='Values')

# Handling outliers by capping
for column in columns_to_plot:
    Q1 = df2[column].quantile(0.25)
    Q3 = df2[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Capping the lower and upper outliers
    df2[column] = df2[column].clip(lower=lower_bound, upper=upper_bound)

# Use the adjusted function to create a melted DataFrame for the selected columns
df2_melted_selected = melt_dataframe(df2, columns_to_plot)

# Create boxplot
plt.figure(figsize=(12, 6))
ax = sns.boxplot(x='Columns', y='Values', data=df2_melted_selected, color="gray")
plt.xticks(rotation=45)
plt.title('Boxplots of Selected Numeric Columns After Handling Outliers')

# Removing grid lines
ax.grid(False)

# Removing the top and right border lines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Keeping only the bottom and left axis lines, making them minimalistic
ax.spines['bottom'].set_visible(True)
ax.spines['left'].set_visible(True)

# Remove x-axis and y-axis tick marks
ax.tick_params(axis='both', length=0)

# Ensuring labels are present on both the x-axis and y-axis
ax.set_xlabel('Columns', fontsize=12)  # Customize the font size as needed
ax.set_ylabel('Values', fontsize=12)  # Customize the font size as needed

plt.show()
```

```python
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Define the number of folds for cross-validation
num_folds = 5

# Initialize a KFold object with the desired number of folds
kf = KFold(n_splits=num_folds, shuffle=True, random_state=42)

# Initialize lists to store metrics for each fold
mse_scores = []
r2_scores = []

# Initialize the linear regression model
model = LinearRegression()

# Perform k-fold cross-validation
for train_index, val_index in kf.split(x2):
    # Split the data into training and validation sets for this fold
    x2_train, x2_val = x2.iloc[train_index], x2.iloc[val_index]
    y2_train, y2_val = y2.iloc[train_index], y2.iloc[val_index]
    
    # Fit the model on the training data
    model.fit(x2_train, y2_train)
    
    # Predict the target variable on the validation data
    y2_pred = model.predict(x2_val)
    
    # Calculate the mean squared error for this fold
    mse = mean_squared_error(y2_val, y2_pred)
    mse_scores.append(mse)
    
    # Calculate the R-squared for this fold
    r2 = r2_score(y2_val, y2_pred)
    r2_scores.append(r2)

# Calculate the average MSE and R-squared across all folds
avg_mse = np.mean(mse_scores)
avg_r2 = np.mean(r2_scores)

# Data dictionary containing the results
results_data = {
    "Metric": [
        "Mean Squared Error (MSE) across all folds",
        "Root Mean Squared Error (RMSE)",
        "Average R-squared (R^2) across all folds"
    ],
    "Value": [
        avg_mse,
        np.sqrt(avg_mse),
        avg_r2
    ]
}

# Convert dictionary to DataFrame
results_df = pd.DataFrame(results_data)

# Function to print DataFrame with lines between rows
def print_dataframe_with_lines(df):
    print("-" * 70)  # Print a line at the top
    for index, row in df.iterrows():
        print(f"{row['Metric']}: {row['Value']:.4f}")  # Formatting float to 4 decimal places
        print("-" * 70)  # Print a line after each row

# Display the DataFrame with lines
print_dataframe_with_lines(results_df)
```

```python
# Calculate p-values
from scipy import stats
n = len(y2_train)  # number of samples
p = x2_train.shape[1]  # number of features
dof = max(0, n - p)  # degrees of freedom

# Calculate standard errors of coefficients
mse_resid = np.sum((y2_train - model.predict(x2_train)) ** 2) / dof
se = np.sqrt(np.diagonal(np.linalg.pinv(np.dot(x2_train.T, x2_train)) * mse_resid))

# Calculate t-statistics
t = model.coef_ / se

# Calculate p-values
p_values = 2 * (1 - stats.t.cdf(np.abs(t), dof))
# Organize data into a DataFrame for a neat tabular display
metrics_data = {
    "Feature": x2_train.columns,
    "Coefficient": model.coef_,
    "Standard Error": se,
    "T-Statistic": t,
    "P-Value": p_values
}
metrics_df = pd.DataFrame(metrics_data)


# Function to print the DataFrame with lines
def print_dataframe_with_lines(df):
    # Print column headers with lines
    headers = df.columns
    header_line = ' | '.join(["{:<20}".format(header) for header in headers])
    print(header_line)
    print('-' * (len(header_line) + 10))  # adjust line length based on header length

    # Print each row with lines
    for index, row in df.iterrows():
        row_line = ' | '.join(["{:<20}".format(f"{value:.4f}" if isinstance(value, float) else value) for value in row])
        print(row_line)
        print('-' * (len(header_line) + 10))

# Print the formatted table
print("Regression Analysis Results:")
print_dataframe_with_lines(metrics_df)
```

## Linear Regression

Regression model does not have hyperparameters that require tuning in the same way that models like Decision Tree Regressors or Random Forests do. 

Linear Regression aims to find the coefficients that minimize the residual sum of squares between the observed 
targets in the dataset and the targets predicted by the linear approximation. There are no settings to adjust the
model's complexity or regularization directly within the basic Linear Regression model itself.

```python
from sklearn.linear_model import Ridge

# Define the parameter grid for Ridge Regression
ridge_param_grid = {
    'alpha': np.logspace(-4, 4, 20)  # 20 values between 10^-4 and 10^4
}

# Initialize the Ridge regression model
ridge = Ridge()

# Initialize GridSearchCV with the parameter grid
ridge_grid_search = GridSearchCV(ridge, ridge_param_grid, cv=5, scoring='neg_mean_squared_error')

# Fit GridSearchCV
ridge_grid_search.fit(X_train, y_train)

# Get the best model
best_ridge_model = ridge_grid_search.best_estimator_

# Predict on the testing set with the best model
y_pred = best_ridge_model.predict(X_test)

# Calculate performance metrics
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r_squared = r2_score(y_test, y_pred)



import pandas as pd

# Assuming best_ridge_model, y_test, and y_pred have been properly defined and computed

# Define the results in a dictionary, including the best parameters from GridSearchCV
results_data = {
    "Metric": ["Alpha", "MSE", "RMSE", "R^2"],
    "Value": [
        ridge_grid_search.best_params_['alpha'],  # Alpha parameter
        mse,  # Mean Squared Error
        rmse,  # Root Mean Squared Error
        r_squared  # R-squared
    ]
}

# Create DataFrame from the dictionary
results_df = pd.DataFrame(results_data)

# Function to print the DataFrame with lines for better readability
def print_dataframe_with_lines(df):
    # Print the header
    print(f"{'Metric':<15} | {'Value':>15}")
    print("-" * 33)  # Print a line to separate headers from the data

    # Iterate through DataFrame rows and print each
    for index, row in df.iterrows():
        # Print each row data
        print(f"{row['Metric']:<15} | {format_value(row['Value'])}")
        print("-" * 33)  # Print a line after each row

# Helper function to format different types of values appropriately
def format_value(value):
    if isinstance(value, float):
        return f"{value:.4f}"  # Format floats to 4 decimal places
    return value

# Display the formatted results
print_dataframe_with_lines(results_df)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

# Define a grid of parameters to search
param_grid = {
    'n_estimators': [100, 200, 300],  # Number of trees in the forest
    'max_depth': [None, 10, 20, 30],   # Maximum depth of the trees
    'min_samples_split': [2, 5, 10],   # Minimum number of samples required to split an internal node
    'min_samples_leaf': [1, 2, 4],     # Minimum number of samples required to be at a leaf node
    'max_features': ['auto', 'sqrt', 'log2']  # Number of features to consider when looking for the best split
}

# Initialize GridSearchCV
grid_search = GridSearchCV(RandomForestRegressor(random_state=42), param_grid, cv=5,
                           scoring='neg_mean_squared_error', return_train_score=True)

# Fit GridSearchCV on the training data
grid_search.fit(X_train, y_train)

# Extract the best estimator (model with the best parameters)
best_rf_model = grid_search.best_estimator_

# Predict on the test set with the best model
y_pred_best = best_rf_model.predict(X_test)

# Calculate performance metrics
mse_best = mean_squared_error(y_test, y_pred_best)
rft_rmse_best = np.sqrt(mse_best)
rft_mae_best = mean_absolute_error(y_test, y_pred_best)
rft_r_squared_best = r2_score(y_test, y_pred_best)

# Organize the best parameters and performance metrics into DataFrames
params_df = pd.DataFrame(list(grid_search.best_params_.items()), columns=['Parameter', 'Value'])
metrics_data = {
    'Metric': ['MSE', 'RMSE', 'MAE', 'R^2'],
    'Value': [mse_best, rft_rmse_best, rft_mae_best, rft_r_squared_best]
}
metrics_df = pd.DataFrame(metrics_data)

# Function to print DataFrames with headers and rows formatted
def print_formatted_table(df, title):
    print(title)
    print("-" * (len(df.columns) * 20))  # adjust line length based on number of columns
    header = " | ".join([f"{col_name:<15}" for col_name in df.columns])
    print(header)  # header
    print("-" * (len(df.columns) * 20))  # header underline

    for index, row in df.iterrows():
        row_formatted = []
        for value in row:
            if isinstance(value, float):
                row_formatted.append(f"{value:15.4f}")
            else:
                row_formatted.append(f"{value:15}")  # Use str formatting for non-float values
        print(" | ".join(row_formatted))
    print("-" * (len(df.columns) * 20))  # end line

# Display the best parameters and performance metrics in formatted tables
print_formatted_table(params_df, "Best Model Parameters:")
print_formatted_table(metrics_df, "Performance Metrics:")
```
